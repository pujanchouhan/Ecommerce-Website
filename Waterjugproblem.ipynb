```python
from collections import deque

def water_jug_problem(cap1, cap2, target):
    visited = set()
    queue = deque([(0, 0)])

    while queue:
        x, y = queue.popleft()

        # If target is reached
        if x == target or y == target:
            return True

        # Skip if already visited
        if (x, y) in visited:
            continue

        visited.add((x, y))

        # Generate all possible next states
        queue.extend([
            (cap1, y),  # Fill jug1
            (x, cap2),  # Fill jug2
            (0, y),     # Empty jug1
            (x, 0),     # Empty jug2

            # Pour jug1 -> jug2
            (x - min(x, cap2 - y),
             y + min(x, cap2 - y)),

            # Pour jug2 -> jug1
            (x + min(y, cap1 - x),
             y - min(y, cap1 - x))
        ])

    return False
cap1, cap2, target = 4, 3, 2

if water_jug_problem(cap1, cap2, target):
    print("It is possible to measure", target, "liters.")
else:
    print("It is not possible to measure", target, "liters.")
```
